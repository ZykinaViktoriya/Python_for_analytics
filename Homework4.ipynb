## Задача 3

### Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

Для начала нам необходимо выявить самую высокую корреляцию, чтобы определить зависимость

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, linewidths=1, cmap='coolwarm')
```

```python
plt.figure(figsize=(8,6))
sns.jointplot(data=df, x="sqft_living", y="price", kind="reg", height=12, ratio=4)
plt.title('Зависимость стоимости жилья и жилых кв.метров');
```

Данный график показывает нам прямую зависимость стоимости дома от размера жилой площади

```python
plt.figure(figsize=(8,6))
sns.jointplot(data=df, x="sqft_living15", y="price", kind="reg", height=12, ratio=4)
plt.title('Зависимость стоимости жилья и жилых кв.метров у 15 соседних домов');
```

```python
data = df.groupby('yr_built')['price'].mean().reset_index()
data
```

```python
plt.figure(figsize=(15, 6))
plt.scatter(data['yr_built'], data['price'])
plt.title('Распределение цен в зависимости от года постройки')
plt.xlabel('год строительства дома')
plt.ylabel('стоимость');
```

Распределение цен на жилье в зависимости от года постройки имеет U-образную форму. Пики выявлены в начале 1900-х и в 2000-х, с резким провалом в 1940-х годах.

```python
plt.figure(figsize=(14, 10))

sns.boxplot(x=df['grade'], y=df['price'])

plt.xlabel('Класс дома')
plt.ylabel('Цена,млн')
plt.title('Распределение цен в зависимости от класса дома')
plt.yticks(np.arange(0, 8100000, 500000));
```

С повышением класса дома растет стоимость жилья, диапазон цен и количество выбросов

```python
plt.figure(figsize=(14, 10))
sns.boxplot(x=df['price'], y=df['waterfront'].astype('str'), whis=1.5, hue=df['grade'])

plt.title('Распределение цен по береговой линии')
plt.xlabel('Цена, млн')
plt.ylabel('Наличие набережной')
plt.xticks(np.arange(0, 8100000, 500000));
```

Разброс цен примерно сопоставим для домов имеющих набережную и нет

```python
plt.figure(figsize = (14,10))
sns.boxplot(x = df['price'], y = df['floors'].astype('str'), whis = 1.5)
plt.xlabel('Цена')
plt.ylabel('Этажность')
plt.title('Стоимость/Этажность дома')
plt.xticks(np.arange(0, 8100000, 500000));
```

Наибольший разбег по ценам наблюдается для домов в 2,5 этажей, а минимальный- для домов в 3,5 этажей. По 2-этажным домам наблюдается максимальное количество выбросов, в свою очередь дома с 3,5 этажами почти не имеют выбросов. Самый дорогой дом имеет 2,5 этажа и продается почти за 8 000 000.

```python
plt.figure(figsize = (14,10))
sns.boxplot(x = df['price'], y = df['condition'].astype('str'), whis = 1.5)
plt.xlabel('Цена')
plt.ylabel('Состояние дома')
plt.title('Стоимость/Оценка состояния');
```

Дома, состояние которых оценивается в 1 и 2 почти не имеют выбросов.По домам с состоянием в 3-4 наблюдаются максимальное количество выбросов. Максимальную стоимость имеет дом с оценкой 4(хорошее состояние).
