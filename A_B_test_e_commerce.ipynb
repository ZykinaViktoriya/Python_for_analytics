## Хи квадрат

```python
n1 = df1.query('con_treat == "treatment"').converted.value_counts()[0]
n2 = df1.query('con_treat == "control"').converted.value_counts()[0]
k1 = df1.query('con_treat == "treatment"').converted.value_counts()[1]
k2 = df1.query('con_treat == "control"').converted.value_counts()[1]
```

```python
from statsmodels.stats import proportion
import numpy as np
proportion.proportions_chisquare(np.array([k1, k2]),
                                                   np.array([n1, n2]))
```

## Z test

Стат. значимых различий нет. Нулевую гипотезу не отвергаем

```python
z_score, z_pvalue = proportion.proportions_ztest(np.array([k1, k2]),
                                                   np.array([n1, n2]))
print('Results are ','z_score =%.3f, pvalue = %.3f'%(z_score, z_pvalue))
```
