## Логическое И &
выполняются оба условия одновременно

```python
df['Gender'] == 'Female'
```

```python
df[(df['Gender'] == 'Female') & (df['NumOfProducts'] >= 3)]
```

### Логическое ИЛИ |
выполняется хотя бы одно условие

```python
df[(df['HasCrCard']) | (df['NumOfProducts'] >= 3)]
```

### Логическое НЕ isin/~
true меняется на false

```python
df[~(df['Geography'] == 'Spain')]
```

```python
df[df['Geography'].isin(['France', 'Germany'])]
```

## Индексация

### loc
выводит только явно указанные номера индексов в таблице

```python
df_small = df[(df['Geography'] == 'Spain')][['Geography', 'Gender', 'Age']]
df_small.head()
```

```python
df_small.loc[1]
```

```python
df_small.loc[5]
```

```python
df_small.loc[[1,4,5], ['Gender', 'Age']]
```

### iloc
выводит индексы по порядку

```python
df_small.head()
```

```python
df_small.iloc[[0,1,2]]
```

```python
df_small.iloc[0, [0,2]]
```
