```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Python for analytics/laptop_price.csv', encoding='latin-1')
df.head()
```

### 1.1 Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU

```python
df['Cpu_Company']=df['Cpu'].apply(lambda x: x.split()[0])
df.sample(15)
```

### 1.2 Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя

```python
df['Memory'].value_counts()
```

```python
# 256GB SSD +  1.0TB Hybrid
def memory_size(row):
  row = row.split()
  gb = 0
  for el in row:
    if 'GB' in el:
      gb += int(el[:-2])
    elif 'TB' in el:
      gb += float(el[:-2]) * 1024
  return int(gb)

df['Memory_Amount'] = df['Memory'].apply(memory_size)
df
```

### 1.3 Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD/SDD/др.)

```python
def memory_type(row):
  row = row.split()
  m_type = ''
  for el in row:
    if el.isalpha() and el  not in m_type:
      if m_type:
        m_type += ''
      m_type += el
  return m_type

df['Memory_Type'] = df['Memory'].apply(memory_type)
df

```

```python
def memory_type2(row):
  row = row.split()
  for i in range(len(row)-1,-1,-1):
    if 'GB' in row[i] or 'TB' in row[i]:
     del row[i]
  return ' '.join(row)

df['Memory_Type'] = df['Memory'].apply(memory_type2)
df
```

### 1.4 Удалите признаки Memory и ScreenResolution

```python
df.drop(columns=['Memory', 'ScreenResolution'], inplace=True)
df.sample(10)
```

## Задание 4
### 4.1 Найдите среднюю стоимость ноутбуков в зависимости от компании производителя
Отсортируйте от меньшей стоимости к большей

```python
df.groupby('Company')['Price_euros'].mean().sort_values()
```

```python
df.groupby('Company').agg({'Price_euros': 'mean'}).sort_values('Price_euros')
```

### 4.2 Найдите минимальную, среднюю и максимальную стоимости ноутбуков в зависимости от производителя процессора

```python
grouped = df.groupby('Cpu_Company').agg({'Price_euros':['mean', 'min', 'max']})
grouped
```

### 4.3 Постройте таблицу с подсчетом количества ноутбуков в данных в зависимости от производителя CPU и ОЗУ

```python
df.pivot_table(index='Cpu_Company',
               columns='Ram',
               values='Company',
               aggfunc='count',
               fill_value=0)
```

### 4.4 Постройте таблицу с подсчетом средней стоимости ноутбуков в данных в зависимости от операционной системы и GB памяти

```python
df.groupby(['OpSys', 'Memory_Amount']).agg({'Price_euros': 'mean'}).unstack('OpSys').fillna(0).round(2)
```
